```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image_path):
    """Load image and apply preprocessing steps."""
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    edges = cv2.Canny(image, threshold1=50, threshold2=150)
    return edges

def find_contours(edges):
    """Find contours from the edge-detected image."""
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return contours

def regularize_curve(contour):
    """Simplify the curve using approximation."""
    epsilon = 0.01 * cv2.arcLength(contour, True)
    approx_curve = cv2.approxPolyDP(contour, epsilon, True)
    return approx_curve

def identify_shape(approx_curve):
    """Identify the shape based on the number of vertices and properties."""
    num_vertices = len(approx_curve)

    if num_vertices == 2:
        return "Line"
    elif num_vertices == 3:
        return "Triangle"
    elif num_vertices == 4:
        # Check if it's a rectangle
        angles = []
        for i in range(4):
            pt1 = approx_curve[i][0]
            pt2 = approx_curve[(i + 1) % 4][0]
            pt3 = approx_curve[(i + 2) % 4][0]
            angle = np.arctan2(pt3[1] - pt2[1], pt3[0] - pt2[0]) - np.arctan2(pt1[1] - pt2[1], pt1[0] - pt2[0])
            angles.append(np.abs(angle))

        if all(np.isclose(angle, np.pi/2, atol=0.1) for angle in angles):
            return "Rectangle"
        else:
            return "Quadrilateral"
    elif num_vertices > 4:
        # Check for polygons
        return "Polygon with {} sides".format(num_vertices)
    else:
        return "Unknown"

def fit_ellipse(contour):
    """Fit an ellipse to the contour if possible."""
    if len(contour) >= 5:
        ellipse = cv2.fitEllipse(contour)
        return ellipse
    return None

def is_circle(ellipse):
    """Check if the ellipse is close to a circle."""
    (x, y), (MA, ma), angle = ellipse
    aspect_ratio = ma / MA
    return np.isclose(aspect_ratio, 1.0, atol=0.2)

def detect_shapes(image_path):
    """Main function to process image and detect shapes."""
    edges = preprocess_image(image_path)
    contours = find_contours(edges)

    image = cv2.imread(image_path)
    for contour in contours:
        approx_curve = regularize_curve(contour)
        shape_name = identify_shape(approx_curve)

        # Fit ellipse to contours
        ellipse = fit_ellipse(contour)
        if ellipse:
            if is_circle(ellipse):
                shape_name = "Circle"
            else:
                shape_name = "Ellipse"

        print(f"Shape identified: {shape_name}")

        # Draw the shape on the image
        cv2.drawContours(image, [approx_curve], -1, (0, 255, 0), 2)

    # Show the regularized image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Detected Shapes')
    plt.show()

# Set the image path directly here for Colab
image_path = 'download.png'  # Change this to your image path in Colab
detect_shapes(image_path)
```
